## Project Description

This project is a simple console-based Tic Tac Toe game written in Python.

### Features
- Two-player game (Player X and Player O)
- 3x3 board displayed in the console
- Players take turns choosing positions
- The game checks for:
  - win (row, column, diagonal)
  - draw (no more moves left)
- Input validation for invalid or already taken positions

The goal of this project is to practice:
- basic Python programming
- functions and game logic
- using Jupyter Notebook and GitHub together

```python
# Tic Tac Toe game logic

#Creates Horizontal Lines
def create_board():
    return [" " for _ in range(9)]

#Constructs the Board Design
def print_board(board):
    print()
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print()

#Checks for Win Sequence Horizontally, Vertically, and Diagonally.
def check_winner(board, player):
    # Winning combinations
    wins = [
        (0, 1, 2),
        (3, 4, 5),
        (6, 7, 8),
        (0, 3, 6),
        (1, 4, 7),
        (2, 5, 8),
        (0, 4, 8),
        (2, 4, 6),
    ]
    for a, b, c in wins:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

#Checks Whether the Game Ended in a Draw
def is_draw(board):
    return " " not in board

#Checks if Moves are Valid (Checks Bounds, Open Spaces, Legal Moves)
def make_move(board, position, player):
    if position < 0 or position > 8:
        return False, "Invalid position. Choose 1–9."
    if board[position] != " ":
        return False, "Position already taken."
    board[position] = player
    return True, "Move accepted."
```

```python
def play_game():
    board = create_board()
    current_player = "X"
    print("Welcome to Tic Tac Toe!")
    print("Positions are 1–9, left to right, top to bottom.")
    print_board(board)

    while True:
        try:
            move = int(input(f"Player {current_player}, enter your move (1-9): ")) - 1
        except ValueError:
            print("Please enter a number between 1 and 9.")
            continue

        valid, message = make_move(board, move, current_player)
        if not valid:
            print(message)
            continue

        print_board(board)

        if check_winner(board, current_player):
            print(f"Player {current_player} wins!")
            break

        if is_draw(board):
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"
```

```python
# Run the game

play_game()
```
